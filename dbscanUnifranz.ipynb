```python
from sklearn.preprocessing import StandardScaler ,normalize
```

```python
st = StandardScaler()
```

```python
st_df = st.fit_transform(data)
```

```python
normalizar = normalize(st_df)
```

```python
normalizar
```

```python
normalizacion_df = pd.DataFrame(data = normalizar , columns = data.columns)
```

```python
normalizacion_df
```

```python
from sklearn.cluster import DBSCAN
```

```python
modelo = DBSCAN(eps=0.5,min_samples=8)
```

```python
modelo.fit(normalizacion_df)
```

```python
clusters_labels= modelo.fit_predict(normalizacion_df)
```

```python
data['Clusters'] = clusters_labels
```
